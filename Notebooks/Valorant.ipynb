```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r'..\data\player_stats.csv')
df.head()
```

```python
df.columns
```

```python
df = df[['player', 'agent','map', 'kill', 'assist', 'death', 'kast%']]
df.head()
```

```python
df.nunique()
```

```python
df['agent'].unique()
```

```python
df1 = df.groupby(['player', 'agent', 'map'], as_index=False).sum()
df1
```

```python
df1.isna().sum()
```

```python
print(df1)
```

```python
df1 = df1[df1['map'] == 'Ascent']
df1
```

```python
df1['KDA'] = (df1['kill'] + df1['assist']) / df1['death']
df1
```

```python
valorant = pd.pivot_table(df1, index='player', columns='agent', values='KDA', fill_value=0)
valorant
```

```python
kda = []

for i in valorant.columns:
  data = input(f'Enter the your KDA for {i}: ')
  kda.append(data)

print(kda)
```

```python
from scipy.stats import pearsonr
from sklearn.metrics.pairwise import cosine_similarity

# Define a function to calculate similarity using Pearson correlation coefficient
def pearson_similarity(x, y):
    corr, _ = pearsonr(x, y)
    return corr

# Choose the similarity function: pearson_similarity or cosine_similarity_wrapper
similarity_function = pearson_similarity  # Change this to cosine_similarity_wrapper for Cosine similarity

# Calculate similarity between user's KDA profile and other players' KDA profiles
similarities = valorant.apply(lambda row: similarity_function(row, pd.Series(map(float, kda))), axis=1)

# Find the top 5 most similar players
similar_players = similarities.nlargest(6)[1:]  # Exclude the user itself

# Get the stats of those 5 similar players
similar_players_stats = df1[df1['player'].isin(similar_players.index)]

# Sort the agents according to their KDA among those similar players
sorted_agents = similar_players_stats.groupby('agent')['KDA'].max().nlargest(5)

# Print the recommended agents
print("Top Recommended Agents based on Similar Players' Performance (Sorted by KDA):")
recommended_agents = []
for agent, kda in sorted_agents.items():
    if len(recommended_agents) >= 5:
        break
    if agent not in recommended_agents:
        recommended_agents.append(agent)
        print(f"{agent}: {kda:.2f}")
```
